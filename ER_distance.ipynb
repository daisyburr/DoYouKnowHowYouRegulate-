```python
#create euclidean matrix for each subject
#Euclidean distance goes from 0 to infinity, with smaller numbers representing being closer together
distance = []

for i in range(0,52):
    subject_cor = result[i]
    s_rdm = pdist(subject_cor, 'euclidean')
    distance.append(s_rdm)
    sns.heatmap(squareform(s_rdm), annot = True, cmap='RdYlBu_r', yticklabels=['REAPPRAISE', 'SUPPRESS', 'SPONTANEOUS'],
            xticklabels=['REAPPRAISE', 'SUPPRESS', 'SPONTANEOUS'])
    print(s_rdm) #to print actual distance values
    plt.ylabel('Regulation condition', fontsize=18)
    plt.xlabel('Regulation condition', fontsize=18)
    plt.tight_layout()
    plt.savefig('Distance_Matrix')
    plt.show()



```

```python
distances_df = pd.DataFrame(distance,columns=['REAPPRAISE_SUPPRESS', 'REAPPRAISE_UNINSTRUCTED', 'SUPPRESS_UNISTRUCTED']) #save values into df
df = pd.read_csv('d_neg_avg_ER.csv') #redefine subject
subject = np.unique(df['ID'])#redefine subject
subs = pd.Series(subject)#turn subject into series
distances_df['subject'] = subs.values #then add to df

```

```python
#Correlational distance goes from 0 - 2, with 0 being PERFECT correlation, 1 being no correlation, and 2 being PERFECT ANTICORRELATION.
#correlation distance goes from 0 to 1, with smaller numbers representing being closer together
distance_cor = []

for i in range(0,52):
    subject_cor = result[i]
    s_rdm_cor = pdist(subject_cor, 'correlation')
    distance_cor.append(s_rdm_cor)
    sns.heatmap(squareform(s_rdm_cor), annot = True, cmap='RdYlBu_r', yticklabels=['REAPPRAISE', 'SUPPRESS', 'SPONTANEOUS'],
            xticklabels=['REAPPRAISE', 'SUPPRESS', 'SPONTANEOUS'])
    print(s_rdm_cor) #to print actual distance values
    plt.ylabel('Regulation condition', fontsize=18)
    plt.xlabel('Regulation condition', fontsize=18)
    plt.tight_layout()
    plt.savefig('Distance_Matrix_cor')
    plt.show()

```

```python
distances_df_cor = pd.DataFrame(distance_cor,columns=['REAPPRAISE_SUPPRESS_cor', 'REAPPRAISE_UNINSTRUCTED_cor', 'SUPPRESS_UNISTRUCTED_cor']) #save values into df
df = pd.read_csv('d_neg_avg_ER.csv') #redefine subject
subject = np.unique(df['ID'])#redefine subject
subs = pd.Series(subject)#turn subject into series
distances_df_cor['subject'] = subs.values #then add to df

```
